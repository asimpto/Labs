```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['text.usetex'] = True
import math
import numpy as np
from numpy import diff
```

```python
def f(x):
    return 25 * x ** 3 * (1 - 4 * x ** 2) * (math.e ** (-5 * x ** 2))

def derivative_func(x):
    h = 0.0001
    return (f(x + h) - f(x)) / h

x = np.arange(-1.5, 1.5, 0.01)

y = f(x)

y_diff = derivative_func(x)

y_diff_diff = derivative_func(y_diff)
```

$$
25x^3(1 - 4x^2)e^{-5 \cdot x^2}
$$

```python
plt.plot(x, y)
plt.plot(x, y_diff_diff)
plt.scatter(0, 0, color='red')
#plt.plot(x, y_diff_diff)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["f(x)", "f ''(x)"])
plt.title(r'$25x^3(1 - 4x^2)e^{-5x^2}$')
plt.show()
```
