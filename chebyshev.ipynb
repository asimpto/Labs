```python
import matplotlib.pyplot as plt

def chebyshev_metric(x, y):
    return abs(x) + abs(y)

# Создание точек для отображения окружности
points = []
for x in range(-100, 101):
    for y in range(-100, 101):
        if chebyshev_metric(x, y) == 1:
            points.append((x, y))

# Разделение точек на оси X и Y
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]

# Построение окружности
plt.scatter(x_values, y_values, color='blue', s=5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Единичная окружность в метрике Чебышева')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```
