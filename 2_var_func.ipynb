```python
import matplotlib.pyplot as plt
import numpy as np
import scipy


# Задаем функцию
def f(x1, x2):
    return x1 ** 3 + x2 ** 3 - 3 * x1 * x2

# Создаем массив значений x
x1, x2 = np.meshgrid(np.arange(-1000, 1000, 0.5), np.arange(-1000, 1000, 0.5))

z = f(x1, x2)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x1, x2, z, cmap='inferno')
plt.xlabel("x1")
plt.ylabel("x2")
ax.set_zlabel("z")
plt.show()
solution = scipy.optimize.minimize(fun=f, x0=(0, 0))
```
